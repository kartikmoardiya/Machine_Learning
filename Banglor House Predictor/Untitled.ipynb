```python
def convertRange(x):
    temp = x.split('-')

    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None;
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()

    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft) # location na flat na flate na mean and std have evo j data rakhvano chhe price m-st and m+st ni vachhe hoy
        st = np.std(subdf.price_per_sqft)

        gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index=True)

    return df_output

data = remove_outliers_sqft(data)
data.describe()
```

```python
# bhk na outlier n remove kara
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_sqft),
                'std' : np.std(bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0] # bhk ktla chhe as per location
            }
        print(location, bhk_stats)
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5: # bhk no count 5 to jovu j joye and ene jo 5 thi vadhi jay to ene drop kri nakho
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values) # inde value lay lidhi have ene index na base upar drop kri devani
        print(exclude_indices)
    return df.drop(exclude_indices, axis = 'index')
```

```python
bhk_outlier_remover(data)
```

```python
X = data.drop(columns=['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

### Applying Linear Regression

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output = False), ['location']),
    remainder='passthrough'
)
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

### Applying Lasso


```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler, lasso)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test, y_pred_lasso)
```

### Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_ridge)
```

```python
print("No Regularixation: ", r2_score(y_test,y_pred_lr))
print("Ridge: ", r2_score(y_test,y_pred_lasso))
print("Lasso: ", r2_score(y_test,y_pred_ridge))
```
